```r
source("../Scripts/functions.R")
source("../Scripts/visulizations.r")
library("RColorBrewer")
library(ggplot2)
library(dplyr)
library(purrr)
```

```r
celltypes = c("Cytotoxic T cell", "CD4+ T cell", "CD14+ monocyte", "B cell", "Megakaryocyte", "Natural killer cell",
              "CD16+ monocyte", "Dendritic cell", "Plasmacytoid dendritic cell")
coi <- c("Cytotoxic T cell", "Natural killer cell","Dendritic cell", "Plasmacytoid dendritic cell" )
methods <- c("Seurat",  "SingleR","CellID", "SingleCellNet", "ItClust")
sizes <- c(3090, 2418, 1373, 1022, 703, 623, 273, 126, 23)
names(sizes) <- celltypes

colors <-brewer.pal(length(celltypes), "Paired")
names(colors) <- celltypes
```

```r
data <- read.csv("../Results/Files/results_PBMC10x_long.csv")
data <- data[as.numeric(data$Set) <= 20,]
data_accuracy <- calculate_accuracy(data)
data_accuracy$Approach[data_accuracy$Approach=="SCN"] <- "SingleCellNet"
data_accuracy$Approach <- factor(data_accuracy$Approach, levels=methods)
data_accuracy$class <- factor(data_accuracy$class, levels=celltypes)

data_lineplot <- get_data_lineplot(data_accuracy)


options(repr.plot.width=15, repr.plot.height=10)
f2a = ggplot(data_lineplot[data_lineplot$class %in% coi,], aes(Size, mean, group=Approach)) +
geom_point(size=0.5) + geom_line()+  
geom_ribbon(aes(ymin = p25, ymax = p75), alpha = 0.3)+
facet_grid(cols = vars(class), rows = vars(Approach),  labeller = label_wrap_gen(width=12))+
theme_bw()+ ylab("Accuracy")+ xlab("Cells per cell type")+
geom_hline(aes(yintercept = Accuracy), color="red", linetype="dashed")+
geom_vline(aes(xintercept=refSize), linetype="dotted", color="black")+
theme(axis.text=element_text(size = 6), axis.title=element_text(size = 8),strip.text = element_text(size = 6))+
scale_y_continuous(breaks=c(0,0.5,1))

f2a
```

```r
runtime <- read.csv("../Results//Files//runtime.txt", sep=" ", header=F)
colnames(runtime) <- c("Method", "Tag", "Runtime")
runtime[c('Reference', 'Size', "Set")] <- stringr::str_split_fixed(runtime$Tag, '_', 3)
runtime <- runtime[runtime$Reference =="PBMC10x",]
runtime$Runtime <- as.numeric(runtime$Runtime)
runtime$Size <- as.numeric(runtime$Size)
runtime <- runtime[runtime$Size <= 3000,]

runtime$Method[runtime$Method=="SCN"] <- "SingleCellNet"
runtime$Method <- factor(runtime$Method, levels=methods)
runtime <- runtime[!is.na(runtime$Runtime),]
runtime_summary <- runtime %>% 
           dplyr::group_by(Size, Method, Reference) %>% 
           dplyr::summarize(mean = mean(Runtime),
                            p25 = quantile(Runtime, probs = c(0.25)),
                            p75 = quantile(Runtime, probs = c(0.75))) 

f2b <- ggplot(runtime_summary, aes(as.numeric(Size), mean, color=Method, group=Method, fill= Method)) + 
geom_point() + geom_line() + theme_bw()+ 
geom_ribbon(aes(ymin = p25, ymax = p75), alpha = 0.1)+
ylab("Runtime in seconds") + xlab("Number of cells per celltype") + 
theme(legend.position = "bottom")+
theme(axis.text=element_text(size = 6), axis.title=element_text(size = 6),
      legend.text=element_text(size = 6), legend.title=element_text(size = 6),legend.key.size = unit(0, 'lines'), legend.direction="vertical")+
 guides(linetype=guide_legend(ncol=2),
         color=guide_legend(ncol=2)) 
f2b
```

```r
confidence_scores <- read.csv("../Results/Files/results_PBMC10x_confidencescores_2.csv")
```

```r
confidence_scores$celltype <- factor(confidence_scores$celltype, levels=celltypes)

confidence_scores$method[confidence_scores$method=="SCN"] <- "SingleCellNet"
confidence_scores$method <- factor(confidence_scores$method, levels=methods)

f2c <- ggplot(confidence_scores[confidence_scores$celltype %in% c("Cytotoxic T cell", "Natural killer cell",
                                                                  "Dendritic cell", "Plasmacytoid dendritic cell")
                               & confidence_scores$size %in% c(38,100,1000),],
              aes(as.factor(size), confidence, color=is_correct)) + 
geom_boxplot(alpha=0.3, width=1, position = position_dodge(0.5),  outlier.size=0.1)+
#geom_violin(position=position_dodge(0.5), alpha=0)+
facet_grid(cols=vars(celltype), rows=vars(method),  labeller = label_wrap_gen(width=12))+
theme_bw()+
theme(legend.position = "bottom",
      axis.text=element_text(size = 6), axis.title=element_text(size = 8),
      strip.text = element_text(size = 6),
     legend.text=element_text(size = 6), legend.title=element_text(size = 6), legend.key.size = unit(0, 'lines')) +
xlab("Cells per cell type")+ylab("Confidence Score")+labs(color="Correct label")+ylim(0,1)

f2c
```

```r
s1 <- ggplot(confidence_scores[confidence_scores$celltype %in% celltypes
                               & confidence_scores$size %in% c(38,100,1000),],
              aes(as.factor(size), confidence, color=is_correct)) + 
geom_boxplot(alpha=0.3, width=1, position = position_dodge(0.5),  outlier.size=0.1)+
#geom_violin(position=position_dodge(0.5), alpha=0)+
facet_grid(cols=vars(celltype), rows=vars(method),  labeller = label_wrap_gen(width=12))+
theme_bw()+
theme(legend.position = "bottom",
      axis.text=element_text(size = 6), axis.title=element_text(size = 8),
      axis.text.x=element_text(angle = 25, hjust = 1), strip.text = element_text(size = 6),
     legend.text=element_text(size = 6), legend.title=element_text(size = 6), legend.key.size = unit(0, 'lines')) +
xlab("Cells per cell type")+ylab("Confidence Score")+labs(color="Correct label")+ylim(0,1)

s1
```
